<h1>Data Cleaning</h1>

```python
# First make a copy for original set

d = cdf.copy()
```

```python
# To drop the unwanted column
#  EG from the dataset, Drop the trainsaction id as it is not neccessary to include for analysis

d.drop(['Transaction_ID'],inplace=True,axis=1)
```

```python
d
```

```python
# rename columns

d.rename(columns={'State_names':'State'},inplace=True)
```

```python
# Adding new column
d['Amount_spent_cent'] = 0
```

```python
d['Amount_spent_cent'] = d['Amount_spent']*100
```

```python
d
```

```python
# To change the value in column using condition
d.loc[df.Marital_status=='Single','Marital_status']='S'
```

```python
d
```

```python
d.loc[df.Marital_status=='Married','Marital_status']='M'
```

```python
d
```

```python
d.loc[df.Age<=18,'Age']='Under 18'
```

```python
d
```

```python
d.loc[(df.Age>18) & (df.Age<25),'Age']='18 to 25'
```

```python
d.loc[(df.Age>25) & (df.Age<55),'Age']='25 to 45'
```

```python
d.info()
```

```python
d.duplicated()
```

```python
d.duplicated().sum()
```

```python
d.drop_duplicates(inplace=True, keep='first')
```

```python
# Finding number of missing values and sort them
d1.isna().sum().sort_values(ascending=False)
```

<h2>Input Missing values(NaN)</h2>

```python
d2 = df.copy()
```

```python
d2.isna().sum().sort_values(ascending=False)
```

```python
#impute age=NaN with mode
age_mode=d2['Age'].mode()
age_mode[0]
```

```python
d2['Age'].fillna(age_mode[0],inplace=True)
```

```python
d2.isna().sum()
```

<h2>Memory Management</h2>

```python
# Change object datatype to category
object_columns = d2.select_dtypes(include='object').columns
d2[object_columns] = d2[object_columns].astype('category')
```

```python
d2.info()
```
