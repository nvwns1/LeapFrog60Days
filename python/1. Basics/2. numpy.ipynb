<h1>Numpy</h1>

<h2>Introduction</h2>

* NumPy is a Python library used for working with arrays. <br>
* The array of object in NumPy is called ndarray. <br>
* NumPy stands for Numerical Python.

<h3>Why use numpy</h3>

* Has array object which is 50x faster than Python list <br>
* functions for working in domain of linear algebra and matrices <br>

<h3>Why numpy is faster?</h3>

* NumPy arrays are stored at one continous place in memory unlike lists ( localityy of reference ) <br>
* It is written in C or C++ for fast computation

<h2>Import NumPy</h2>

```python
import numpy as np
```

<h2>Create numPy Arrays</h2>

```python
zero_array = np.zeros((2,3))
ones_array = np.ones((2,3))
random_array = np.random.randint(low=0, high=255, size=(2,3))


print(zero_array)
print(ones_array)
print(random_array)
```

<h2>Accessing and Slicing</h2>

```python
random_array[0][0]
```

```python
random_array[-1][-1]
```

```python
print(random_array)
print(random_array[0:2])
print(random_array[0:2][-1])
random_array[0:2][-1][0:2]
```

<h2>Manipulation</h2>

```python
np.insert(random_array, 1, 5)
```

```python
np.delete(random_array, 2)
```

<h2>Arithmetic</h2>

```python
array1 = np.array([[1,2,3]])
array2 = np.array([[4,5,6]])
addition = np.add(array1, array2)
print(addition)
print(np.subtract(array1, array2))
print(np.multiply(array1, array2))
print(np.divide(array1, array2))
```

<h2>Aggregation Functions</h2>

```python
print(random_array)
print(np.sum(random_array))
print(np.mean(random_array))
print(np.std(random_array))
print(np.median(random_array))
print(np.max(random_array))
print(np.min(random_array))
```

<h2>Reshaping</h2>

```python
reshaped_array = random_array.reshape(3,2)
print(reshaped_array)
```

<h2>Broadcasting</h2>

```python
scalar =  2
result = reshaped_array * scalar
print(result)
```
