<h1>Pandas</h1>

* Pandas is a powerful and open-source python library. <br>
* It is build on top of the NumPy library. <br>
* It is well-suited for working with tabular data such as spreadsheets and SQL tables.<br>
* Pandas are generally used for data science.

<h2 style="color:skyBlue;">What can Pandas do?</h2>

* Data set cleaning, merging and joining <br>
* Easy handling of missing dta (represented as NaN) in floating points as well as non-floating point data.<br>
* Columns can be inserted and deleted from Dataframe and higher dimensional objects.<br>
* Powerful group by functionality for performing split-spply-combine operations on data sets. <br>
* Data visualization

<h2 style="color: Skyblue;">Basic data structures in pandas</h2><br>
Pandas provides two types of classes for handling data.<br>

1. Series: <br>
    a. one-dimensional label array holding data of any type( integers, strings, Python, object, etc)

2. Dataframe: <br>
    a. two-dimensionla data structure that holds data like a two-dimension array or a table with a rows and column


<h3 style="color: skyblue">Pandas Series</h3>

* A Pandas Series os ;ole a column in a table.

```python
import pandas as pd

# Simple array
data = [1, 2, 3, 4, 5]

ser = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])

print(ser)
print("The type of pandas series is: ", type(ser))
```

<h3 style="color: skyblue">Data Frames </h3>

* DataFrame is like a multi-dimensional table <br>
* Series is a column and a Dataframe is the whole table.

```python
import pandas as pd

data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data)
print(df)
print("The type of pandas dataframe is: ", type(df))
print(df.head())

```

<h4>Notes</h4>

* Pandas use the loc attribute to return one or more specified rows

```python
print(df.loc[1])
print(df.loc[[0, 1]])
```

<h4 style="color:skyblue">Replace Empty values</h4>

* The fillna() method allows us to replace empty cells with a value


```python
df.fillna(20)
```

```python
import pandas as pd

df = pd.read_csv('data/fitness.csv')
df.isna().sum()
df.fillna(20, inplace=True) # fill all missing values with 20
df.isna().sum()
```

<h2 style="color:skyblue">Cleaning data of Wrong format</h2>

* Data of wrong format makes it difficult/ impossible to analyze data
* There are two option to fix the data of wrong format

1. Remove the rows <br>
2. Convert into similar format

<h3>Removing Rows</h3>

```python
import pandas as pd

df = pd.read_csv('data/fitness.csv')
df.head()

df1 = df.copy()

df.dropna(subset=['Maxpulse'], inplace=True)
 # drop rows with missing values in the Maxpulse column
```

<h3>Replacing Values</h3>


```python
df.head(6)
```

```python
df.loc[6, 'Duration'] = 45 # replace the value in the Duration column at index 6 with 45
df.loc[6]
```

<h3>Removing Rows</h3>

```python
import pandas as pd

df = pd.read_csv('data/fitness.csv')

for x in df.index:
    if df.loc[x, 'Duration'] > 120:
        df.drop(x, inplace=True)

#  Remember to include the 'inplace = True' argument to make the changes in the original dataframe


print(df.to_string())
```
