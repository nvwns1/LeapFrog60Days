<h1>Deep Learning</h1>

A subset of machine learning that focuses on using neural networks with many layers (hence "Deep") to model and understand complex patterns in data. It is particulary effective for tasks involving large amounts of data and is a key technology behind many recent advances in artificial intelligence.

<h2>Key Concepts</h2>

1. Artificial Neural Networks:
The programs designed to solve any problem try to mimic the structure and the function of our nervous system. They recieve input, process it and pass it to the next layer.

2. Layer: Composed of neurons. There are different types of layers.
* Input Layer: Recieves the raw data.
* Hidden Layer: Intermediate layers where computation occurs. The term "deep" refers to the presence of multiple hidden layers
* Output Layer: Produces the final prediction or classification.

3. Activation function: Functions applied to each neuron;s ouput to introduce the non-linearity, allowing the network to learn complex patterns.
* Sigmoid
* ReLU (Rectified Linear Unit)
* Tanh

4. Backpropagation
 A technique for training the network by adjusting weights based on the error of prediction.
* Loss Function
* Gradient Descent




<h2>General Terminologies</h2>

* Weight: It determines how much the current input affects the output. The machine finds the weight through the training process. It is a parameter
* Bais: It is a constant which helps the model in way that it can fit best for the given data. Bias is a constant which gives freedom to perform best

<h2>Types of Neural Network</h2>

-  Perception
- Feed Forward Neural Networks
- Convolutional Neural Networks
- Recurrent Neural Networks
- Long Short Term Memory (LSTM) Networks

<h3>Perception</h3>

- Simplest and oldest model of neuron
- Takes input, sum the, up and apply activation Function 


<h3>Feed Forward</h3>

1. Architecture: 
* Consists of an input layer, one or multiple hidden layers and an output layer.
* All nodes are fully connected

2. Use Caases: Basic classification and regression tasks

3. Example: Predicting house price based on features like size, location and number of rooms.


<h3>Recurrent Neural Network</h3>

1. Architecture: Includes connections that form directed cycles, allowing information to persist across the steps. Use hidden states to maintain context.

2. Use Cases: Sequential data processing, such as time series prediction and NLP

3. Example: Predicting the next word in a sentences


<h3>Long Short Term Memory </h3>

1. Architecture: A Type of RNN with special units called memory cells that can maintain information for long periods. Includes gates to control the flow of information

2. Use Cases: Long-term dependencies in sequence data,

3. Example: Generating text based on an initial prompt.



<h3>Convulutional Neural Networks</h3>

1. Archtecture: 
* Composed of convolutional layers, pooling layers and fully connected layers. Convolutional layer apply filters to input data to detect features
* Convolutional Layer and pooling layer capture low level details of the images.

2. Use Cases: Image and video analysis, object detection and facial recognition

3. Example: Recognizing objects in an image


<h2>Activation Function</h2>

* The mathematical equations that determines the output of a neural network.
It normalize the output of each neuron to a certain range like tanh activation function normalizes between -1 to 1.
* Also known as Transfer Function

<h3>Types of Activation Function</h3>

1. Linear Activation Function: 
* A straight line function activation is proportional to input. 
* It is of form : A = cx
* Range: (-infinity to infinity)

Disadvantage
* Not possible to use back propagation to train the model
* Limited power and ability to handle complexity varying parameters of input data

2. Non-Linear Activation function
* Those which have more than one and they have a curvature when we plot.
* It makes it easy for the model to generalize or adapt with variety of data.

<h4>Types of Non-Linear Activation Function</h5>

* Sigmoid function
* Tanh function
* ReLu function
* Leaky ReLu
* Softmax function

<h5>Sigmoid Function</h5>

* It is a activation function of form f(x) = 1/1+exp(-x)
* Range (0,1)
* Use Cases: Binary Classification problems, output layer oif binary classifier



<h5>Tanh function</h5>

* f(x) = 1 - exp(-2x)/1 + exp(-2x)
* Range (1, -1)
* Use Cases: Hidden layers of neural networks where the centering of data is important.




<h5>ReLU (Rectified Linear Unit)</h5>

* ReLU(x) = max(0, x)
* Range: [0, infinity]
* Use Cases: Hidden layers of neural network, especially in CNN and deep networks.



<h5>Leaky ReLu</h5>

* Imporved Version of ReU Function
* Leaky ReLU(x) = max(ax, x) where a is a small constant (usually 0.01)
* Range: (-infinity, infinity)
* Use Cases: Hidden Layers where the Dying ReLU problem is observed



<h5>Softmax Function</h5>

* Range(0,1) for ach class with the sum of all outputs equal to 1
* Use cases: Output layer of neural networks for multi-class classification

<h3>Which Activation Function is the best</h3>

1. For Hidden Layers in Deep networks:
* ReLu is often good default choice due to its computational effieciency and ability to mitigate the vanishing gradient problem
* If you encounter dying ReLUs (neurons that always ouput zero) consider Leaky ReLU.\

2. For Output Layers:
* Use Sigmoid for binary classification
* Use Softmax for multi-class classification


* Note:
Hyperbolic tan function: Tanh function is mostly used in recurrent neural network for natural language processing and speech recognition task 

<h2>Loss Functions for Neural Networks</h2>

* Loss functions helps in optimizing the parameters of the neural networks.
* Our objective is to minimize the loss for a neural network by optimizing its parameter (weights).
* The loss is calculated using loss function by matchint the target value and predicted value by a neural network

We Use the gradient descent method to optimize the weights of the network such that loss is minimized.

<h3>Different Loss Functions</h3>

* Mean Squared Error (L2 Loss)
* Corss Entropy Losss
* L1 Loss

<h3>Mean Squared Error (L2 Loss)</h3>

* We apply back propagation algorithm to iteratively minimise the MSE.
* Use Cases: Penalize larger errors more significantly than smaller ones.



<h3>Cross Entropy Loss</h3>

* It is more advanced mean squared error commonly used in classification or regression problems. 

<h4>Binary Cross Entropy</h4>

- Used when you have a binary classification task. Using BCe loss function, just need one output node to classify the data into two classes.
- The output value should be passed through a sigmoid activation function so the output is in the range (0-1)

<h4>Categorical Cross Entropy</h4>

- Used when you have a multi-class classification task
- Using CCE Loss function, there must be the same number of output nodes as the classes.
- The final layer output should be passed through a softmax activation so that each node output a probabilty value between (0,-1) 


<h3>Gradient Descent</h3>

Gradient Descent is an optimization algorith for finding a local minimum of a differentiable function.

<h4>key Concepts</h4>

1. Loss function: A function that measures the difference between the predicted values and the actual values. The goal of trainin g a model is to minimize the loss function.
2. Gradient: The vector of partial derivatives of loss function with respect to the model's parameter. It indicates the direction and rate of the steepestt increase in the loss of function
3. Learning Rate: A hyperparameter that determines the step size of each iteration while moving towards the minimum. 



<h4>Types of Gradient Descent</h4>

1. Batch Gradient Descent
* Uses the entire training dataset to comput the gradient at each step
* Pros: Converges to the global minimul for convex loss functions.

2. Stochastic Gradient Descent
* Uses one training example to compute the gradient and update the parameters at each step
* Pros: Faster and can escape local minima
* Cons: More noisy and less stable convergence

3. Mini-Batch Gradient Descent
* Uses a small random subset of the trianing data to compute the gradient at each step
* Pros: balances the efficiency of batch gradient descent and the speed of SGD.
* Cons: Requires choosing the mini batch size
