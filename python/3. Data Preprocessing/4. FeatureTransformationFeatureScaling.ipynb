```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('data/Titanic_ML.csv', usecols=['Age', 'Fare', 'Survived'])
data.head()
```

<h2>Logarithmic Transformation</h2>

```python
data['Age_log'] = np.log(data['Age'])
plt.figure(figsize=(10, 7))
sns.displot(data['Age_log'])
plt.show()
```

```python
data['fare_log'] = np.log1p(data['Fare'])
plt.figure(figsize=(10, 7))
sns.displot(data['fare_log'])
plt.show()
```
